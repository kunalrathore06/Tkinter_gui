```python
from tkinter import ttk
from tkinter import filedialog
from tkinter import *
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import seaborn as sns
from matplotlib.figure import Figure
import tkinter as tk

root = Tk()
root.title("DashBoard")
root.rowconfigure(0, weight=1)
root.columnconfigure(0, weight=1)
root.state('zoomed')

frame1 = LabelFrame(root, text="Data")
frame1.place(height=300,width=430)

frame2 = LabelFrame(root, text="Bar Chart")
frame2.place(height=300,width=430,rely=0.00,relx=0.34)

frame3 = LabelFrame(root, text="Scatter Chart")
frame3.place(height=300,width=430,rely=0.00,relx=0.68)

frame4 = LabelFrame(root, text="Line Chart")
frame4.place(height=300,width=430,rely=0.50,relx=0.34)

file_frame = LabelFrame(root,text="open file")
file_frame.place(height=100,width=400,rely=0.65,relx=0.00)

b1 = Button(file_frame,text="Browse a File",command=lambda:fill_dialog(),bg="#20bebe", fg="white")
b1.place(rely=0.65,relx=0.00)

b2 = Button(file_frame,text="Load a File",command=lambda:load_data(),bg="#20bebe", fg="white")
b2.place(rely=0.65,relx=0.30)


label_file = ttk.Label(file_frame,text="No file selected")
label_file.place(rely=0.00,relx=0.00)

label=Label(root, text="", font=("Courier 22 bold"))
label.grid()

X_axis_title = tk.Label(root, text="X Axis").place(rely=0.80,relx=0.70)
X_axis = tk.StringVar()
X_axisEntry = tk.Entry(root, textvariable=X_axis,width= 40).place(rely=0.80,relx=0.80)  

Y_axis_title = tk.Label(root,text="Y Axis").place(rely=0.85,relx=0.70)  
Y_axis = tk.StringVar()
Y_axisEntry = tk.Entry(root, textvariable= Y_axis,width= 40).place(rely=0.85,relx=0.80) 

#Create a Button to validate Entry Widget
Button(root, text= "Plot",width= 20, command= lambda:submit(X_axis,Y_axis)).place(rely=0.90,relx=0.85)
# ttk.Button(root, text= "Y-axis",width= 20, command= display).place(rely=0.90,relx=0.80)

def Barchart(file_name,X_axis, Y_axis):
    figure1 = plt.Figure(figsize=(4.5,3), dpi=100)
    df1 = pd.read_csv(file_name)
    ax1 = figure1.add_subplot(111)
    bar1 = FigureCanvasTkAgg(figure1, master = frame2)
    bar1.get_tk_widget().grid()
    df1 = df1[Y_axis].groupby(df1[X_axis]).sum()
    df1.plot(kind='bar', legend=True, ax=ax1)
    ax1.set_title('Bar Chart')

    
def Scatterchart(file_name,X_axis, Y_axis):
    df3 = pd.read_csv(file_name)
    figure3 = plt.Figure(figsize=(4.5,3), dpi=100)
    ax3 = figure3.add_subplot(111)
    ax3.scatter(df3[X_axis],df3[Y_axis], color = 'g')
    scatter3 = FigureCanvasTkAgg(figure3, master = frame3) 
    scatter3.get_tk_widget().grid()
    ax3.set_title("Scatter Chart")

def Linechart(file_name,X_axis, Y_axis):
    df2 = pd.read_csv(file_name)
    figure2 = plt.Figure(figsize=(4.5,3), dpi=100)
    ax2 = figure2.add_subplot(111)
    line1 = FigureCanvasTkAgg(figure2, master = frame4)
    line1.get_tk_widget().grid()
    df2 = df2[Y_axis].groupby(df2[X_axis]).sum()
    df2.plot(kind='line', legend=True, ax=ax2, color='r',marker='o', fontsize=10)
    ax2.set_title('Line Chart')


def submit(X_axis, Y_axis ):
    X_axis= X_axis.get()
    Y_axis= Y_axis.get()
     
    Barchart(file_name,X_axis, Y_axis)
    Linechart(file_name,X_axis, Y_axis)
    Scatterchart(file_name,X_axis, Y_axis)
    

tv1 = ttk.Treeview(frame1)
tv1.place(relheight=1,relwidth=1)
treescrolly = Scrollbar(frame1,orient="vertical",command=tv1.yview)
treescrollx = Scrollbar(frame1,orient="horizontal",command=tv1.xview)
tv1.configure(xscrollcommand = treescrollx.set,yscrollcommand=treescrolly.set)
treescrollx.pack(side="bottom",fill="x")
treescrolly.pack(side="right",fill="y")

tv2 = ttk.Treeview(frame2)
tv2.place(relheight=0.9,relwidth=1)

tv3 = ttk.Treeview(frame3)
tv3.place(relheight=0.9,relwidth=1)

tv4 = ttk.Treeview(frame4)
tv4.place(relheight=0.9,relwidth=1)



def fill_dialog():
    global file_name
    file_name = filedialog.askopenfilename(initialdir=":/",
                                          title="select a file",
                                          filetypes=[("CSV files", ".csv")])
    label_file['text'] = file_name
    return None

def load_data():
    file_path = label_file['text']
    try:
        csv_filename=r"{}".format(file_path)
        df = pd.read_csv(csv_filename)
        
    except ValueError:
        messagebox.showerror("Imformation","The upload file is invalid")
        return None
    except filenotfound:
        messagebox.showerror("Imformation",f"No such file as {file_path}")
        return None
    
    clear_data()
    
    tv1["column"] = list(df.columns)
    tv1["show"] = "headings"
    for column in tv1["column"]:
        tv1.heading(column,text=column)
    df_row = df.to_numpy().tolist()
    for row in df_row:
        tv1.insert("","end",values=row)

def clear_data():
    tv1.delete(*tv1.get_children())

    

    

root.mainloop()
```
